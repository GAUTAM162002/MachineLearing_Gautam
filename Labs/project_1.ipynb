```python
import pandas as pd
import matplotlib.pyplot as plt
gapminder_data = pd.read_csv(r"C:\Users\harsh\OneDrive\Desktop\GIT HUB\MachineLearing_Gautam\datasets\gapminderDataFiveYear.csv")
gapminder_data
```

```python
European_countries = ['France', 'United Kingdom', 'Italy','Germany', 'Spain']
```

```python
European_countries_df = gapminder_data[gapminder_data['country'].isin(European_countries)]
```

```python
plt.figure(figsize=(10,6))
```

```python


for country in European_countries:
    country_data = European_countries_df[European_countries_df['country'] == country]
    plt.plot(country_data['year'], country_data['lifeExp'], marker='o', label=country)
plt.title('Life Expectancy in European Countries Over Time')
plt.xlabel('year')
plt.ylabel('Life Expectancy')
plt.legend(title='Country')
plt.grid()
plt.show()
```
