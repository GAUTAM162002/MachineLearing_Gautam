##### The Planets dataset contains details about the 1,000+ extrasolar planets discovered up to 2014. Visualize the distribution of the masses of the planets (expressed as a multiple of the mass of Jupiter) using a histogram and a box plot. Make appropriate modifications to the chart title, axis titles, legend, figure size, font size, colors, etc., to make the chart readable and visually appealing.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mass_value= pd.read_csv(r"C:\Users\harsh\OneDrive\Desktop\GIT HUB\MachineLearing_Gautam\datasets\planets.csv")
mass_value
```

```python
mass_value = mass_value["mass"].dropna()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.hist(mass_value,
         bins=30,
         color="skyblue",
         edgecolor="black"
)


plt.title("PlanetMass Ditribution")
plt.xlabel("Count")
plt.ylabel("mass(Jupiter Masses)")
```

```python


plt.subplot(1,2,2)
plt.boxplot(mass_value,vert =False)
plt.title("Planet Mass Box Plot")
plt.xlabel("Mass(Jupiter Masses)")
plt.tight_layout()
plt.show()
```
