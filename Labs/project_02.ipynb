```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r'C:\Users\harsh\OneDrive\Desktop\GIT HUB\MachineLearing_Gautam\datasets\diamonds.csv')
df
```

```python
df = df[(df["clarity"]=="SI2")&(df["color"]=="E")]
```

```python
plt.figure(figsize=(8,5))
```

```python
sc =plt.scatter(
    df["carat"],
    df["price"],
    c=df["cut"].astype("category").cat.codes,
    cmap="viridis",
    alpha=0.7
)
plt.xlabel("Carat")
plt.ylabel("Price")
plt.title("Carat vs Price (Colord by Cut)")
plt.colorbar(sc,label="Cut category")
plt.show()
```
